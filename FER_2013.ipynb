```python
# =============================================================================
# OPTIMIZED FER2013 PIPELINE WITH PERFORMANCE RECOVERY
# -----------------------------------------------------------------------------
# This script implements a complete Convolutional Neural Network (CNN) pipeline
# for Facial Expression Recognition (FER) using the challenging FER2013 dataset.
# It includes optimized data processing, a robust CNN architecture, and an
# improved training strategy with regularization and imbalance mitigation.
# =============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, callbacks, optimizers, regularizers
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
# Note: imblearn's RandomOverSampler is imported but not explicitly used in the 'smart' strategy.
# The 'smart' balancing is implemented manually below.
import cv2
import warnings
import os
from google.colab import drive
from collections import Counter

# Suppress minor warnings for a cleaner output, common in ML pipelines.
warnings.filterwarnings('ignore')

# Set random seeds for NumPy and TensorFlow to ensure the training results
# are reproducible across different runs.
np.random.seed(42)
tf.random.set_seed(42)

# =============================================================================
# 1. OPTIMIZED DATA PROCESSING
# -----------------------------------------------------------------------------
# This class handles loading the dataset, parsing the pixel data, applying
# essential image preprocessing, and splitting the data for training and testing.
# =============================================================================

class OptimizedFERProcessor:
    def __init__(self, csv_path):
        # Path to the FER2013 CSV file.
        self.csv_path = csv_path
        # Map integer emotion labels to human-readable names for reporting.
        self.emotion_map = {
            0: 'Angry', 1: 'Disgust', 2: 'Fear',
            3: 'Happy', 4: 'Sad', 5: 'Surprise', 6: 'Neutral'
        }

    def load_data(self):
        """Load and analyze data with better validation"""
        print("Loading FER2013 dataset...")
        self.df = pd.read_csv(self.csv_path)
        # Add emotion names for easier analysis and plotting.
        self.df['emotion_name'] = self.df['emotion'].map(self.emotion_map)

        print(f"Dataset shape: {self.df.shape}")

        # Display the crucial emotion distribution, highlighting the severe class
        # imbalance (e.g., Disgust vs. Happy). This justifies the need for
        # class weighting later.
        print("\n=== EMOTION DISTRIBUTION ===")
        emotion_counts = self.df['emotion_name'].value_counts().sort_index()
        for emotion, count in emotion_counts.items():
            percentage = (count / len(self.df)) * 100
            print(f"{emotion}: {count} samples ({percentage:.2f}%)")

        return self.df

    def optimized_preprocessing(self, images):
        """Applies essential preprocessing for better feature visibility."""
        print("Preparing images with optimized preprocessing...")
        # Contrast Limited Adaptive Histogram Equalization (CLAHE) is used.
        # This technique enhances contrast, especially in low-light/low-resolution
        # areas, which is critical for the 48x48 FER2013 images.
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        processed = []

        for img in images:
            # Convert float image back to 8-bit integer (0-255) for OpenCV processing.
            img_uint8 = (img * 255).astype(np.uint8)

            # Apply CLAHE.
            img_processed = clahe.apply(img_uint8)

            # Re-normalize the image back to a float range of 0.0 to 1.0.
            processed.append(img_processed.astype(np.float32) / 255.0)

        return np.array(processed)

    def prepare_datasets(self, test_size=0.15, val_size=0.15, balance_strategy='smart'):
        """Splits data and handles class imbalance using smart oversampling."""

        # 1. Pixel Parsing
        pixels = []
        for pixel_str in self.df['pixels']:
            # Convert the space-separated pixel string into a 48x48 NumPy array.
            img_array = np.array(pixel_str.split(), dtype=np.float32).reshape(48, 48)
            pixels.append(img_array)

        pixels = np.array(pixels) / 255.0  # Initial normalization
        pixels = self.optimized_preprocessing(pixels) # Apply CLAHE
        # Reshape to (samples, height, width, channels) format for Keras CNN.
        pixels = pixels.reshape(-1, 48, 48, 1)

        y = self.df['emotion'].values

        # 2. Split into Training/Validation/Test Sets
        # Stratified split ensures all emotion classes are represented proportionally
        # in the train and test sets, which is vital due to the imbalance.
        X_temp, X_test, y_temp, y_test = train_test_split(
            pixels, y, test_size=test_size, stratify=y, random_state=42, shuffle=True
        )

        # 3. Smart Oversampling (Addressing Imbalance)
        if balance_strategy == 'smart':
            # Oversampling is applied only to the training portion (X_temp/y_temp).
            # It samples up to 150% of the median class count to avoid overfitting
            # to synthetic duplicates of minority classes.

            class_counts = np.bincount(y_temp)
            # Target count is 1.5 times the median class count.
            target_counts = np.median(class_counts[class_counts > 0]) * 1.5

            X_balanced_list = [X_temp]
            y_balanced_list = [y_temp]

            for class_idx in range(len(self.emotion_map)):
                # Identify minority classes (those below the target count)
                if class_counts[class_idx] > 0 and class_counts[class_idx] < target_counts:
                    # Select random indices (with replacement) to duplicate samples.
                    class_mask = (y_temp == class_idx)
                    X_class = X_temp[class_mask]

                    n_needed = int(target_counts - len(X_class))
                    if n_needed > 0:
                        indices = np.random.choice(len(X_class), n_needed, replace=True)
                        X_balanced_list.append(X_class[indices])
                        y_balanced_list.append(np.full(n_needed, class_idx))

            X_balanced = np.vstack(X_balanced_list)
            y_balanced = np.hstack(y_balanced_list)
        else:
            X_balanced, y_balanced = X_temp, y_temp

        # Final validation split (stratified on the balanced set)
        val_ratio = val_size / (1 - test_size)
        X_train, X_val, y_train, y_val = train_test_split(
            X_balanced, y_balanced, test_size=val_ratio, stratify=y_balanced, random_state=42
        )

        # 4. Conservative Class Weights
        # These weights are calculated based on the inverse frequency of each class
        # in the training set, giving higher weight to minority classes (like Disgust)
        # during loss calculation.
        class_weights = compute_class_weight(
            'balanced', classes=np.unique(y_train), y=y_train
        )
        # Conservative clipping ensures no single class dominates the training,
        # preventing the model from becoming too aggressive on minority classes.
        class_weights = np.clip(class_weights, 0.7, 3.0)
        class_weights = dict(enumerate(class_weights))

        print(f"\n=== OPTIMIZED DATASET SPLITS ===")
        print(f"Training set: {X_train.shape}")
        print(f"Validation set: {X_val.shape}")
        print(f"Test set: {X_test.shape}")
        print(f"Conservative class weights: {class_weights}")

        return (X_train, y_train), (X_val, y_val), (X_test, y_test), class_weights

# =============================================================================
# 2. SIMPLIFIED BUT EFFECTIVE MODEL ARCHITECTURE
# -----------------------------------------------------------------------------
# This section defines the Convolutional Neural Network (CNN) used for feature
# extraction and classification. The architecture is a VGG-like sequence of
# Conv blocks, specifically designed for small (48x48) grayscale images.
# =============================================================================

def create_optimized_model(input_shape=(48, 48, 1), num_classes=7):
    """Creates a deep CNN optimized for FER2013."""

    model = keras.Sequential([
        # First block: 48x48 resolution, 32 filters
        layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape),
        layers.BatchNormalization(), # Stabilizes learning and speeds up convergence
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2), # Reduces feature map size to 24x24
        layers.Dropout(0.25), # Regularization to prevent overfitting

        # Second block: 24x24 resolution, 64 filters
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2), # Reduces feature map size to 12x12
        layers.Dropout(0.25),

        # Third block: 12x12 resolution, 128 filters
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2), # Reduces feature map size to 6x6
        layers.Dropout(0.25),

        # Fourth block: 6x6 resolution, 256 filters
        layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),

        # Dimensionality Reduction and Feature Aggregation
        layers.GlobalAveragePooling2D(), # Reduces 6x6x256 tensor to a 256-element vector
        layers.Dropout(0.5), # High dropout before the dense layers

        # Classifier (Dense Layers)
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(64, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.2),

        # Output Layer: 7 units (one for each emotion) with Softmax activation
        # to output probabilities.
        layers.Dense(num_classes, activation='softmax')
    ])

    return model

# =============================================================================
# 3. IMPROVED TRAINING STRATEGY
# -----------------------------------------------------------------------------
# This class defines the compilation and training process, including crucial
# callbacks for robust performance and overfitting prevention.
# =============================================================================

class OptimizedTrainer:
    def __init__(self, model, learning_rate=0.001):
        self.model = model
        self.learning_rate = learning_rate
        self.history = None

    def compile_model(self):
        """Sets up the optimizer, loss function, and metrics."""
        # Adam optimizer is used with default parameters.
        optimizer = optimizers.Adam(
            learning_rate=self.learning_rate,
            beta_1=0.9,
            beta_2=0.999
        )

        self.model.compile(
            optimizer=optimizer,
            # Used for multi-class classification where labels are integers (0-6).
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )

    def get_optimized_callbacks(self, checkpoint_path):
        """Defines callbacks for robust training and saved model management."""
        return [
            # Early Stopping: Prevents overfitting by monitoring validation accuracy.
            # Training stops if validation accuracy doesn't improve significantly
            # (min_delta=0.001) for 15 epochs (patience=15).
            callbacks.EarlyStopping(
                monitor='val_accuracy',
                patience=15,
                restore_best_weights=True, # Restores weights from the epoch with highest val_accuracy.
                verbose=1,
                min_delta=0.001
            ),
            # ReduceLROnPlateau: Adjusts learning rate dynamically.
            # If validation loss plateaus (doesn't decrease) for 8 epochs,
            # the learning rate is halved (factor=0.5). This helps the model
            # escape local minima towards the end of training.
            callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=8,
                min_lr=1e-7,
                verbose=1
            ),
            # ModelCheckpoint: Saves the model with the highest validation accuracy
            # encountered during training.
            callbacks.ModelCheckpoint(
                checkpoint_path,
                monitor='val_accuracy',
                save_best_only=True,
                verbose=1
            ),
            # CSVLogger: Records training metrics (loss, accuracy, val_loss, val_accuracy)
            # to a CSV file for post-analysis and plotting.
            callbacks.CSVLogger('optimized_training_history.csv')
        ]

    def train_model(self, X_train, y_train, X_val, y_val,
                    class_weights=None, epochs=100, batch_size=32):
        """Executes the training process."""

        self.compile_model()

        print(f"\n=== OPTIMIZED TRAINING ===")
        print(f"Training samples: {len(X_train)}")
        print(f"Validation samples: {len(X_val)}")
        print(f"Batch size: {batch_size}")

        self.history = self.model.fit(
            X_train, y_train,
            validation_data=(X_val, y_val),
            epochs=epochs,
            batch_size=batch_size,
            callbacks=self.get_optimized_callbacks('best_optimized_model.keras'),
            # Applies the calculated weights to the loss function, mitigating
            # the impact of the class imbalance.
            class_weight=class_weights,
            verbose=1,
            shuffle=True
        )

        return self.history

# =============================================================================
# 4. COMPREHENSIVE ANALYSIS TOOLS
# -----------------------------------------------------------------------------
# This section contains utilities for evaluating the trained model's performance,
# generating plots, and detailing error patterns.
# =============================================================================

class PerformanceAnalyzer:
    def __init__(self, emotion_map):
        self.emotion_map = emotion_map

    def analyze_performance(self, model, X_test, y_test):
        """Runs the main evaluation sequence on the test set."""
        print("\n" + "="*60)
        print("PERFORMANCE ANALYSIS")
        print("="*60)

        # Standard loss and accuracy calculation.
        test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
        print(f"Test Accuracy: {test_accuracy:.4f}")
        print(f"Test Loss: {test_loss:.4f}")

        # Generate raw probability predictions and convert to discrete labels.
        y_pred_proba = model.predict(X_test, verbose=0)
        y_pred = np.argmax(y_pred_proba, axis=1)

        # Classification report: Provides per-class Precision, Recall, and F1-score.
        print("\n=== CLASSIFICATION REPORT ===")
        print(classification_report(y_test, y_pred,
                                     target_names=self.emotion_map.values(),
                                     digits=4))

        self.plot_confusion_matrix(y_test, y_pred)

        # Detailed analysis covers per-class performance, top errors, and confidence.
        self.detailed_analysis(y_test, y_pred, y_pred_proba)

        return test_accuracy, y_pred

    def plot_confusion_matrix(self, y_true, y_pred):
        """Generates visualizations of the raw count and normalized confusion matrices."""
        cm = confusion_matrix(y_true, y_pred)
        # Normalized CM shows the Recall (diagonal elements are the recall score).
        cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

        # Plot 1: Raw counts (shows number of misclassifications).
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax1,
                    xticklabels=self.emotion_map.values(),
                    yticklabels=self.emotion_map.values())
        ax1.set_title('Confusion Matrix (Counts)', fontsize=14, fontweight='bold')

        # Plot 2: Normalized values (shows percentage of correct/incorrect predictions).
        sns.heatmap(cm_normalized, annot=True, fmt='.3f', cmap='Blues', ax=ax2,
                    xticklabels=self.emotion_map.values(),
                    yticklabels=self.emotion_map.values())
        ax2.set_title('Confusion Matrix (Normalized)', fontsize=14, fontweight='bold')

        plt.tight_layout()
        plt.show() # Display the plots.

    def detailed_analysis(self, y_true, y_pred, y_pred_proba):
        """Prints a detailed breakdown of metrics and error patterns."""
        from sklearn.metrics import precision_recall_fscore_support

        # Recalculate P/R/F1 for printing.
        precision, recall, f1, support = precision_recall_fscore_support(y_true, y_pred, average=None)

        print(f"\n=== DETAILED PERFORMANCE ANALYSIS ===")
        # Print a formatted table of per-emotion metrics.
        print(f"{'Emotion':<12} {'Precision':<10} {'Recall':<10} {'F1-Score':<10}")
        print("-" * 45)

        for i, emotion in self.emotion_map.items():
            print(f"{emotion:<12} {precision[i]:<10.4f} {recall[i]:<10.4f} {f1[i]:<10.4f}")

        self.error_analysis(y_true, y_pred)
        self.confidence_analysis(y_true, y_pred, y_pred_proba)

    def error_analysis(self, y_true, y_pred):
        """Identifies and reports the 10 most common misclassification pairs."""
        errors = y_pred != y_true
        error_indices = np.where(errors)[0]

        if len(error_indices) > 0:
            # Create pairs of (True Label, Predicted Label) for all errors.
            error_pairs = [(y_true[i], y_pred[i]) for i in error_indices]
            # Use Counter to find the most frequent pairs.
            common_errors = Counter(error_pairs).most_common(10)

            print(f"\n=== TOP ERROR PATTERNS ===")
            for (true, pred), count in common_errors:
                true_name = self.emotion_map[true]
                pred_name = self.emotion_map[pred]
                # Reports errors like "True: Fear -> Pred: Sad: 179 errors"
                print(f"True: {true_name} -> Pred: {pred_name}: {count} errors")

    def confidence_analysis(self, y_true, y_pred, y_pred_proba):
        """Examines the model's prediction certainty."""
        # Confidence is the maximum probability outputted by the softmax layer.
        confidences = np.max(y_pred_proba, axis=1)

        print(f"\n=== CONFIDENCE ANALYSIS ===")
        print(f"Mean confidence: {np.mean(confidences):.4f}")
        print(f"Confidence std: {np.std(confidences):.4f}")

        # Analyze confidence for correct vs. incorrect predictions.
        correct_conf = confidences[y_pred == y_true]
        incorrect_conf = confidences[y_pred != y_true]

        # Ideally, correct confidence should be high and incorrect confidence low.
        if len(correct_conf) > 0:
            print(f"Correct predictions confidence: {np.mean(correct_conf):.4f}")
        if len(incorrect_conf) > 0:
            print(f"Incorrect predictions confidence: {np.mean(incorrect_conf):.4f}")

# =============================================================================
# 5. MAIN OPTIMIZED PIPELINE
# -----------------------------------------------------------------------------
# Orchestrates the entire process from data loading to final evaluation and saving.
# =============================================================================

def run_optimized_pipeline(csv_path, retrain=True):
    """The main execution function for the performance recovery/full training pipeline."""
    print("=" * 60)
    print("OPTIMIZED FER2013 PIPELINE")
    print("=" * 60)

    # Mount Google Drive (necessary if running in a Colab environment to access data/models).
    try:
        drive.mount('/content/drive')
    except:
        print("Google Drive already mounted or not in Colab environment")

    # Step 1: Data preparation
    print("\n1. OPTIMIZED DATA PREPARATION")
    processor = OptimizedFERProcessor(csv_path)
    processor.load_data()

    (X_train, y_train), (X_val, y_val), (X_test, y_test), class_weights = \
        processor.prepare_datasets(balance_strategy='smart')

    # Step 2: Model creation/loading
    print("\n2. MODEL SETUP")
    model_path = '/content/drive/MyDrive/fer_training/best_optimized_model.keras'

    if retrain or not os.path.exists(model_path):
        # Create a new model if retraining is requested or no model is found.
        print("Creating new optimized model...")
        model = create_optimized_model()
        model.summary()

        # Step 3: Training
        print("\n3. OPTIMIZED TRAINING")
        trainer = OptimizedTrainer(model)
        history = trainer.train_model(
            X_train, y_train, X_val, y_val, class_weights, epochs=100, batch_size=32
        )
    else:
        # Load a previously trained model for evaluation.
        print("Loading existing model...")
        model = keras.models.load_model(model_path)
        history = None # No training history available if loaded.

    # Step 4: Evaluation
    print("\n4. COMPREHENSIVE EVALUATION")
    analyzer = PerformanceAnalyzer(processor.emotion_map)

    # Always load the best saved model weights (from ModelCheckpoint) for evaluation.
    try:
        best_model = keras.models.load_model('best_optimized_model.keras')
        print("Restoring model weights from the end of the best epoch...")
    except:
        # Fallback to the current model if the checkpoint wasn't saved (e.g., in quick eval).
        best_model = model

    test_accuracy, y_pred = analyzer.analyze_performance(best_model, X_test, y_test)

    # Plot training history if training actually occurred.
    if retrain and history:
        plot_training_history(history)

    # Final summary
    print("\n" + "=" * 60)
    print("PIPELINE SUMMARY")
    print("=" * 60)
    print(f"Final Test Accuracy: {test_accuracy:.4f}")

    # Compare against a known baseline for the dataset (0.6295 is common).
    baseline_accuracy = 0.6295
    improvement = test_accuracy - baseline_accuracy
    print(f"Improvement over baseline: {improvement:+.4f}")

    if improvement > 0:
        print("✅ Performance recovered successfully!")
    else:
        print("⚠️ Further optimization needed")

    # Save the final best model.
    best_model.save('final_optimized_model.keras')
    print("Model saved as 'final_optimized_model.keras'")

    return best_model, test_accuracy

# =============================================================================
# Helper function for plotting
# =============================================================================

def plot_training_history(history):
    """Plots the training and validation accuracy and loss over epochs."""
    try:
        # Handle potential differences in Keras history keys ('acc' vs 'accuracy').
        if 'accuracy' in history.history:
            acc_key = 'accuracy'
            val_acc_key = 'val_accuracy'
        else:
            acc_key = 'acc'
            val_acc_key = 'val_acc'

        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

        # Plot 1: Accuracy
        ax1.plot(history.history[acc_key], label='Training Accuracy')
        ax1.plot(history.history[val_acc_key], label='Validation Accuracy')
        ax1.set_title('Model Accuracy')
        ax1.legend()
        ax1.grid(True, alpha=0.3)

        # Plot 2: Loss
        ax2.plot(history.history['loss'], label='Training Loss')
        ax2.plot(history.history['val_loss'], label='Validation Loss')
        ax2.set_title('Model Loss')
        ax2.legend()
        ax2.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Warning: Could not plot training history: {e}")

# =============================================================================
# 6. QUICK EVALUATION FUNCTION
# -----------------------------------------------------------------------------
# Allows for rapid testing of a pre-trained model without time-consuming training.
# =============================================================================

def quick_evaluation(csv_path):
    """Performs a fast evaluation on the test set using a pre-trained model."""
    print("=" * 60)
    print("QUICK EVALUATION")
    print("=" * 60)

    # Load processor and data
    processor = OptimizedFERProcessor(csv_path)
    processor.load_data()

    # Prepare test set only (no balancing needed for test set).
    (_, _), (_, _), (X_test, y_test), _ = processor.prepare_datasets(balance_strategy='none')

    # Search for and load the best available model.
    model_paths = [
        '/content/drive/MyDrive/fer_training/best_optimized_model.keras',
        '/content/drive/MyDrive/fer_training/best_enhanced_fer_model.keras',
        'best_optimized_model.keras'
    ]

    model = None
    for path in model_paths:
        if os.path.exists(path):
            try:
                model = keras.models.load_model(path)
                print(f"Loaded model from: {path}")
                break
            except:
                continue

    if model is None:
        print("❌ No suitable model found for evaluation")
        return

    # Evaluate the loaded model.
    analyzer = PerformanceAnalyzer(processor.emotion_map)
    accuracy, _ = analyzer.analyze_performance(model, X_test, y_test)

    print(f"\nQuick Evaluation Accuracy: {accuracy:.4f}")
    return accuracy

# =============================================================================
# 7. EXECUTION
# -----------------------------------------------------------------------------
# Entry point of the script, handling user input to select the operational mode.
# =============================================================================

if __name__ == "__main__":
    # Define the expected path to the dataset CSV file.
    csv_path = '/content/drive/MyDrive/fer2013.csv'

    print("Select mode:")
    print("1. Full training pipeline (retrain model)")
    print("2. Quick evaluation only")
    print("3. Performance recovery (recommended)")

    try:
        # Prompt user for their choice.
        choice = int(input("Enter choice (1-3): ").strip())
    except:
        choice = 3  # Default to performance recovery if input is invalid.

    if choice == 1:
        # Full retraining mode.
        model, accuracy = run_optimized_pipeline(csv_path, retrain=True)
    elif choice == 2:
        # Quick evaluation mode.
        accuracy = quick_evaluation(csv_path)
    else:
        # Performance recovery mode (Full pipeline with focus on optimization).
        print("\n🚀 RUNNING PERFORMANCE RECOVERY PIPELINE")
        model, accuracy = run_optimized_pipeline(csv_path, retrain=True)
```
